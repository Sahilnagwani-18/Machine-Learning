```python
import pandas as pd
import numpy as np
import re
df = pd.read_csv("faa.csv")

print("First 5 rows:")
display(df.head())

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("Original shape:", df.shape)
print("Original missing counts:")
print(df.isnull().sum())
```

```python
df['EVENT_LCL_TIME'] = df['EVENT_LCL_TIME'].astype(str).str.replace('Z', '', regex=False).str.strip()
df['EVENT_LCL_DATE'] = df['EVENT_LCL_DATE'].astype(str).str.strip()

df['EVENT_DATETIME'] = pd.to_datetime(df['EVENT_LCL_DATE'] + ' ' + df['EVENT_LCL_TIME'],
                                      errors='coerce', infer_datetime_format=True)
print("\nSample EVENT_DATETIME (first 5):")
display(df[['EVENT_LCL_DATE', 'EVENT_LCL_TIME', 'EVENT_DATETIME']].head())
```

```python
df['ENTRY_DATE'] = pd.to_datetime(df['ENTRY_DATE'], errors='coerce', infer_datetime_format=True)

# ---------- Select required columns ----------
cols_needed = [
    'ACFT_MAKE_NAME',   # 1. Aircraft make name
    'LOC_STATE_NAME',   # 2. State name
    'ACFT_MODEL_NAME',  # 3. Aircraft model name
    'RMK_TEXT',         # 4. Text information
    'FLT_PHASE',        # 5. Flight phase (may be missing)
    'EVENT_TYPE_DESC',  # 6. Event description type
    'FATAL_FLAG',       # 7. Fatal flag
    'ACFT_DMG_DESC'     # damage description (for one-hot)
]
```

```python
cols_present = [c for c in cols_needed if c in df.columns]
df_req = df[cols_present].copy()
print("\nColumns retained:", df_req.columns.tolist())
print("Shape after subsetting:", df_req.shape)
```

```python
print("\nUnique FATAL_FLAG values before fill:", df_req['FATAL_FLAG'].unique() if 'FATAL_FLAG' in df_req else None)
```

```python
if 'FATAL_FLAG' in df_req:
    # Replace NaN with 'No' (as you intended). Use vectorized fillna.
    df_req['FATAL_FLAG'] = df_req['FATAL_FLAG'].fillna('No')
    # Normalize strings: strip and upper/lower as you prefer. We'll standardize to 'YES'/'NO'
    df_req['FATAL_FLAG'] = df_req['FATAL_FLAG'].astype(str).str.strip().str.upper()
    # Some datasets use 'Y'/'N' or other values; make them consistent:
    df_req['FATAL_FLAG'] = df_req['FATAL_FLAG'].replace({'Y': 'YES', 'N': 'NO', 'NA': 'NO', 'NONE': 'NO'})

print("FATAL_FLAG value counts after fill:")
print(df_req['FATAL_FLAG'].value_counts(dropna=False))
```

```python
if 'FLT_PHASE' in df_req:
    if df_req['FLT_PHASE'].dropna().empty:
        print("\nFLT_PHASE has no non-null values; skipping mode fill.")
    else:
        flt_phase_mode = df_req['FLT_PHASE'].dropna().mode()[0]
        df_req['FLT_PHASE'] = df_req['FLT_PHASE'].fillna(flt_phase_mode)
        print("\nFilled missing FLT_PHASE with mode:", flt_phase_mode)
```

```python
if 'ACFT_DMG_DESC' in df_req:
    if df_req['ACFT_DMG_DESC'].dropna().empty:
        print("ACFT_DMG_DESC has no non-null values; skipping mode fill.")
    else:
        acft_dmg_mode = df_req['ACFT_DMG_DESC'].dropna().mode()[0]
        df_req['ACFT_DMG_DESC'] = df_req['ACFT_DMG_DESC'].fillna(acft_dmg_mode)
        print("Filled missing ACFT_DMG_DESC with mode:", acft_dmg_mode)
```

```python
print("\nMissing counts after filling FATAL_FLAG, FLT_PHASE, ACFT_DMG_DESC:")
print(df_req.isnull().sum())
```

```python
before_rows = len(df_req)
df_req = df_req.dropna(subset=['ACFT_MAKE_NAME'])
after_drop_rows = len(df_req)
print(f"\nDropped rows missing ACFT_MAKE_NAME: {before_rows - after_drop_rows} rows dropped. New shape: {df_req.shape}")
```

```python
missing_pct_threshold = 0.75
min_non_null = int((1 - missing_pct_threshold) * len(df_req))
print(f"\nDropping columns with less than {min_non_null} non-null values (>{int(missing_pct_threshold*100)}% missing).")
```

```python
cols_before = df_req.columns.tolist()
df_req = df_req.dropna(axis=1, thresh=min_non_null)
cols_after = df_req.columns.tolist()
dropped_cols = set(cols_before) - set(cols_after)
print("Dropped columns due to missingness:", dropped_cols)
print("Columns remaining:", df_req.columns.tolist())
```

```python
print("\nNumber of observations after cleaning:", len(df_req), " (before cleaning was", df.shape[0], ")")
```

```python
if 'ACFT_MAKE_NAME' in df_req:
    acft_counts = df_req['ACFT_MAKE_NAME'].value_counts()
    print("\nAircraft counts (how many times each type appears):")
    display(acft_counts)
```

```python
if 'FATAL_FLAG' in df_req:
    fatal_yes = df_req[df_req['FATAL_FLAG'].str.strip().str.upper() == 'YES']
    print("\nObservations with Fatal Flag == 'YES':", len(fatal_yes))
    display(fatal_yes)
```

```python
if 'ACFT_DMG_DESC' in df_req:
    # Clean ACFT_DMG_DESC strings a bit
    df_req['ACFT_DMG_DESC'] = df_req['ACFT_DMG_DESC'].astype(str).str.strip().replace({'nan': np.nan})
    # Create dummies and drop first to avoid multicollinearity
    damage_dummies = pd.get_dummies(df_req['ACFT_DMG_DESC'], prefix='DMG', drop_first=True)
    df_onehot = pd.concat([df_req.reset_index(drop=True), damage_dummies.reset_index(drop=True)], axis=1)
    print("\nOne-hot encoded damage columns added:", damage_dummies.columns.tolist())
else:
    df_onehot = df_req.copy()
    print("\nACFT_DMG_DESC not present — skipped one-hot encoding.")
```

```python
phase_keywords = ['LANDING', 'TAKEOFF', 'CRUISE', 'APPROACH', 'CLIMB', 'DESCENT', 'TAXI', 'GROUND', 'FINAL']
```

```python
def extract_flight_phase_from_text(txt):
    if pd.isnull(txt):
        return 'UNKNOWN'
    txtu = str(txt).upper()
    # find first keyword occurrence in the order of phase_keywords
    for kw in phase_keywords:
        if kw in txtu:
            return kw
    return 'UNKNOWN'
```

```python
df_onehot['FLIGHT_PHASE_TEXT'] = df_onehot['RMK_TEXT'].apply(extract_flight_phase_from_text)

```

```python

if 'FLT_PHASE' in df_onehot:
    # Normalize both for comparison (upper, strip)
    cmp_df = df_onehot[['FLT_PHASE', 'FLIGHT_PHASE_TEXT']].copy()
    cmp_df['FLT_PHASE_NORM'] = cmp_df['FLT_PHASE'].astype(str).str.strip().str.upper().replace({'NAN': np.nan})
    cmp_df['FLIGHT_PHASE_TEXT_NORM'] = cmp_df['FLIGHT_PHASE_TEXT'].astype(str).str.strip().str.upper()
    # Create contingency table
    ctab = pd.crosstab(cmp_df['FLT_PHASE_NORM'].fillna('MISSING'), cmp_df['FLIGHT_PHASE_TEXT_NORM'], margins=True)
    print("\nCrosstab of existing FLT_PHASE vs inferred FLIGHT_PHASE_TEXT:")
    display(ctab)
    # Compute simple matching rate where FLT_PHASE is not missing
    valid = cmp_df['FLT_PHASE_NORM'].notna()
    if valid.any():
        matches = (cmp_df.loc[valid, 'FLT_PHASE_NORM'] == cmp_df.loc[valid, 'FLIGHT_PHASE_TEXT_NORM']).sum()
        match_rate = matches / valid.sum()
        print(f"Match rate between FLT_PHASE and FLIGHT_PHASE_TEXT (on non-missing FLT_PHASE): {match_rate:.2%} ({matches}/{valid.sum()})")
    else:
        print("No non-missing FLT_PHASE values to compare.")
```
